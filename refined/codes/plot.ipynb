```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.colors as mcolors

# Define a custom colormap with strong contrast and white at zero
cmap = mcolors.LinearSegmentedColormap.from_list("custom_cmap", [("darkblue"), ("white"), ("darkred")])
norm = mcolors.TwoSlopeNorm(vmin=np.min(growth_rates), vcenter=0, vmax=np.max(growth_rates))
plt.figure(figsize=(8, 6))
ax = sns.heatmap(growth_rates.T, xticklabels=xi_0, yticklabels=r_omega_values, cmap=cmap, annot=True, fmt=".2f", norm=norm)

plt.xlabel(r'$\xi_0$', fontsize=14)
plt.ylabel(r'$R_\Omega$', fontsize=14)
plt.title('Growth Rate Heatmap', fontsize=16)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Create a heatmap plot with e folding time
e_folding_time = abs(1 / growth_rates)

plt.figure(figsize=(8, 6))
sns.heatmap(e_folding_time.T, xticklabels=xi_0, yticklabels=r_omega_values, cmap="viridis", annot=True, fmt=".2f")
plt.xlabel(r'$\xi_0$', fontsize=17)
plt.ylabel(r'$R_\Omega$', fontsize=17)
plt.title('Growth Rate Heatmap', fontsize=17)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
contour = plt.contourf(xi_0, r_omega_values, e_folding_time.T, levels=20, cmap="viridis")
cbar = plt.colorbar(contour)
cbar.set_label(r'E-folding Time ($\tau$)', fontsize=14)
plt.xlabel(r'$\xi_0$', fontsize=17)
plt.ylabel(r'$R_\Omega$', fontsize=17)
plt.title('E-folding Time Contour Plot', fontsize=17)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(9, 7))
ax = fig.add_subplot(111, projection='3d')
X, Y = np.meshgrid(xi_0, r_omega_values)
ax.plot_surface(X, Y, e_folding_time.T, cmap='viridis', edgecolor='k')

ax.set_xlabel(r'$\xi_0$', fontsize=14)
ax.set_ylabel(r'$R_\Omega$', fontsize=14)
ax.set_zlabel(r'E-folding Time ($\tau$)', fontsize=14)
ax.set_title('E-folding Time 3D Surface', fontsize=16)
plt.show()




```

```python
R_omega=-20
R_k=0.3
h=1
A=0.155
B=0.079
eta_0=1
xi_0=1
eta=eta_0
z=0
```

```python
def alp_m():
    alp_m = (-2*eta_0*np.exp(-2*z**2/h**2)*xi_0*R_omega/(R_k*np.pi*h))*(-A*xi_0 + B)
    return alp_m
```

```python
#for i in range 0 to +5, plot alp_m
z = np.linspace(-1, 1, 100)
alp_m = alp_m()
plt.plot(z, alp_m)
plt.xlabel(r'$\xi_0$')
plt.ylabel(r'$\alpha_m$')
plt.title(r'$\alpha_m$ vs $\xi_0$')
plt.show()



```

```python
def growth_rate():

    alp_m = (-2*eta_0*np.exp(-2*z**2/h**2)*xi_0*R_omega/(R_k*np.pi*h))*(-A*xi_0 + B)
    gamma = np.sqrt(R_omega*np.pi*alp_m/(2*h))-eta_0*np.pi**2/(4*h**2)
    return gamma
    
#for z in range -1 to 1, plot gowth rate vs xi_0

z = np.linspace(0, 5, 100)
gamma = growth_rate()
plt.plot(z, gamma)
plt.xlabel(r'$\xi_0$')
plt.ylabel(r'$\gamma$')
plt.title(r'$\gamma$ vs $\xi_0$')
plt.show()

```
